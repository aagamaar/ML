Importing the dependencies

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Iris DataSet

```python
#Loading the dataset to pandas DataFrame
bc_data = pd.read_csv("/content/breast-cancer.csv")
```

```python
bc_data.head()
```

```python
# Finding the count of different labels
bc_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(bc_data.diagnosis)
```

```python
#appending the labels to the dataframe
bc_data['target'] = labels
```

```python
bc_data.head()
```

0 --------> Benign

1 --------> Malignant

```python
bc_data['target'].value_counts()
```
